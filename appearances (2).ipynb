```python
# Importing all the required libraries
import numpy as np
import pandas as pd
import pymongo
from pymongo import MongoClient

# Reading the csv file
Fifa = pd.read_csv('D:/appearances.csv')
Fifa.head()
```

```python
# Checking the total null values
Fifa.isnull().sum()
```

```python
# Dropping all the 'na' values
Fifa_clean = Fifa.dropna()
Fifa_clean
```

```python
# calculating the total null values after removing all the null values
Fifa_clean.isnull().sum()
```

```python
print(len(Fifa)) # Printing the length of rows in Fifa
print(len(Fifa_clean)) # Printing the length of rows in Fifa_clean
```

```python
# Sorting the dataset with date in ascending order
Fifa_clean.sort_values(by='date',ascending=True)
```

```python
# Renaming the column names:
Fifa_clean.rename(columns={"player_position":"position"},inplace=True)
Fifa_clean.rename(columns={"yellow_cards":"yellowCards"},inplace=True)
Fifa_clean.rename(columns={"red_cards":"redCards"},inplace=True)
Fifa_clean.rename(columns={"player_name":"name"},inplace=True)
Fifa_clean.rename(columns={"home_club_goals":"homeGoals"},inplace=True)
Fifa_clean.rename(columns={"away_club_goals":"awayGoals"},inplace=True)
Fifa_clean
```

```python
# Dropping the columns which are not required
Fifa_final = Fifa_clean.drop(columns=['home_club_id','away_club_id','player_id','player_club_id','game_id','appearance_id'],axis=1)
Fifa_final
```

```python
#creating dict() to send to mongo
Fifa_dict = Fifa_final.to_dict('records')

#making connection
client = pymongo.MongoClient('mongodb://localhost:27017/')
dblist = client.list_database_names()

#creating a database
db = client['Fifa']

#Check if "mydatabase" exists:
if db in dblist:
  print(f"The database {db} exists.")
  #droping the database  
  client.drop_database(db)
else:
    print("New Database added successfully !!")

#creating a collection
appearances = db["appearances"]

#sending dataframe from python to mongo
appearances.insert_many(Fifa_dict)
```

```python
# converting into Data frame
Fifa_test = pd.DataFrame(list(appearances.find()))
```

```python
#deleting the first _id column

del Fifa_test['_id']
Fifa_test
```

```python
import mysql.connector

#establishing the connection
conn = psycopg2.connect(
   user='postgres', password='sachin', host='localhost', database='postgres'
)

#Creating a cursor object using the cursor() method
cursor = conn.cursor()

#Retrieving single row
sql = '''SELECT * from ucl INNER JOIN fifa ON ucl.last_name = fifa.last_name INNER JOIN players ON fifa.last_name = players.last_name'''

#Executing the query
cursor.execute(sql)

#Fetching 1st row from the table
result = cursor.fetchall();
#print(result)

#Closing the connection
conn.close()
```
